## Problem Statement

This project aims to analyze the influence of corn and soybeans, America's 
most farmed corps, on key economic indicators in rural U.S. areas for 
approximately 15 years. By examining crop-related production and sales 
metrics this analysis seeks to identify correlations between crop production 
trends and rural economic development. We will investigate whether corn 
and soybean production data can be used to forecast changes in rural 
employment, income, and poverty levels. The goal is to assess the long-term
economic impacts of crop productivity and sales on rural communities and 
to explore the predictive potential of corn data for future economic 
outcomes.

**Goals**

- *Analyze the influence of corn and soybean production on rural 
economic indicators (employment, income, poverty).*
- *Identify correlations between corn and soybean production metrics 
(e.g., acres planted, yield, production volume, cash receipts) and rural
economic development indicators.*
- *Forecast economic outcomes based on corn and soybean 
production trends (can corn or soybean production be used to predict 
employment, income, and poverty levels?).*
- *Assess the long-term impacts of corn and soybean productivity on 
rural economic outcomes.*

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

---
### Load Data

```python
# Reading in corn data.
corn_acres = pd.read_csv('../data/corn_data/acres_planted.csv')
corn_cash_receipts = pd.read_csv('../data/corn_data/cash_receipts.csv')
corn_price_received = pd.read_csv('../data/corn_data/price_received.csv')
corn_production_bushels = pd.read_csv('../data/corn_data/production_bushels.csv')
corn_production_dollars = pd.read_csv('../data/corn_data/production_dollars.csv')
corn_yield_bushels = pd.read_csv('../data/corn_data/yield_bushels.csv')
```

```python
# Reading in soybean data.
soy_acres = pd.read_csv('../data/soybean_data/acres_planted.csv')
soy_cash_receipts = pd.read_csv('../data/soybean_data/cash_receipts.csv')
soy_price_received = pd.read_csv('../data/soybean_data/price_received.csv')
soy_production_bushels = pd.read_csv('../data/soybean_data/production_bushels.csv')
soy_production_dollars = pd.read_csv('../data/soybean_data/production_dollars.csv')
soy_yield_bushels = pd.read_csv('../data/soybean_data/yield_bushels.csv')
```

```python
# Reading in economic data.
employment = pd.read_csv('../data/economic_data/employment_rural.csv')
personal_income = pd.read_csv('../data/economic_data/personal_income_rural.csv')
poverty_levels = pd.read_csv('../data/economic_data/poverty_levels.csv')
```

### Check Data Types & Null Values

```python
# Making a dictionary for each category for easy looping.

# Corn Data Sets.
corn = {'corn_acres': corn_acres,
       'corn_cash_receipts': corn_cash_receipts,
       'corn_price_received': corn_price_received,
       'corn_production_bushels': corn_production_bushels,
       'corn_production_dollars': corn_production_dollars,
       'corn_yield_bushels': corn_yield_bushels}

# Soy Data Sets.
soy = {'soy_acres': soy_acres,
       'soy_cash_receipts': soy_cash_receipts,
       'soy_price_received': soy_price_received,
       'soy_production_bushels': soy_production_bushels,
       'soy_production_dollars': soy_production_dollars,
       'soy_yield_bushels': soy_yield_bushels}

# Economic Data Sets.
economic = {'employment': employment,
           'personal_income': personal_income,
           'poverty_levels': poverty_levels}
```

```python
# Looping through dictionary to find null values and data types of each data set

# First corn.
for name, df in corn.items():
    # Converting columns to snake_case
    df.rename(columns = lambda col: col.lower().replace(' ', '_'), inplace = True)
    
    print('='*40)
    print(f'Data Set: {name}')
    print(f'Null Count: {df.isna().sum().sum()}')
    print(f'Data Types: \n{df.dtypes}')
```

- **`corn_cash_receipts` missing 144 values.**
- **All data types reflecting the respective values stored.**

```python
# Looping through soy data sets for null values and data types
for name, df in soy.items():
    # Converting columns to snake_case
    df.rename(columns = lambda col: col.lower().replace(' ', '_'), inplace = True)
    
    print('='*40)
    print(f'Data Set: {name}')
    print(f'Null Count: {df.isna().sum().sum()}')
    print(f'Data Types: \n{df.dtypes}')
```

- **`soy_cash_receipts` missing 414 missing values and 2 unnamed columns.**
- **All data types reflecting the respective values stored.**

```python
# Looping through economic data sets for null values and data types
for name, df in economic.items():
    # Converting columns to snake_case
    df.rename(columns = lambda col: col.lower().replace(' ', '_'), inplace = True)
    
    print('='*40)
    print(f'Data Set: {name}')
    print(f'Null Count: {df.isna().sum().sum()}')
    print(f'Data Types: \n{df.dtypes}')
```

- *No missing values found*

### Data Cleanup

```python
# Dropping unnamed columns in soy_cash_receipts

soy_cash_receipts.drop(columns = ['unnamed:_17', 'unnamed:_18'], inplace = True)
```

```python
# Exploring just corn_cash_receipts and soy_cash_receipts for now:

# Setting State column to be index.
corn_cash_receipts.set_index('state', inplace = True)
soy_cash_receipts.set_index('state', inplace = True)
```

```python
# Checking what states are missing in corn_cash_receipts and soy_cash_receipts
missing_corn_states = corn_cash_receipts[corn_cash_receipts.isna().all(axis = 1)].index
missing_soy_states = soy_cash_receipts[soy_cash_receipts.isna().all(axis = 1)].index

print(f'States with no corn cash receipts data: \n{list(missing_corn_states)}')
print('='*100)
print(f'States with no soy cash receipts data: \n{list(missing_soy_states)}')
```

```python
# Filling states with zero data in corn_cash_receipts and soy_cash_receipts

corn_cash_receipts.fillna(0, inplace = True)

soy_cash_receipts.fillna(0, inplace = True)

corn_cash_receipts.isna().sum().sum(), soy_cash_receipts.isna().sum().sum()
```

```python
# Exploring the improper data types.

poverty_levels.dtypes
```

```python
# Converting number_in_poverty to integer by removing commas and type casting
poverty_levels['number_in_poverty'] = poverty_levels['number_in_poverty'].str.replace(',', '').astype(int)
print(f'number_in_poverty column is now {poverty_levels['number_in_poverty'].dtype} type')
```

```python
# Converting total_population to integer by removing commas and type casting
poverty_levels['total_population'] = poverty_levels['total_population'].str.replace(',', '').astype(int)
print(f'total_population column is now {poverty_levels['total_population'].dtype} type')
```

```python
# Some columns which should be renamed in the economic data sets
employment.rename(columns = {'geoname': 'state'}, inplace = True)
personal_income.rename(columns = {'geoname': 'state'}, inplace = True)
```

```python
# Pivoting corn & soy dataframes to have a datetime index and states as columns with the corresponding values

corn_acres_pivot = corn_acres.pivot(index = 'year', columns = 'state', values = 'value')
corn_acres_pivot.index = pd.to_datetime(corn_acres_pivot.index, format = '%Y')

soy_acres_pivot = soy_acres.pivot(index = 'year', columns = 'state', values = 'value')
soy_acres_pivot.index = pd.to_datetime(soy_acres_pivot.index, format = '%Y')

corn_production_dollars_pivot = corn_production_dollars.pivot(index = 'year', columns = 'state', values = 'value')
corn_production_dollars_pivot.index = pd.to_datetime(corn_production_dollars_pivot.index, format = '%Y')

soy_production_dollars_pivot = soy_production_dollars.pivot(index = 'year', columns = 'state', values = 'value')
soy_production_dollars_pivot.index = pd.to_datetime(soy_production_dollars_pivot.index, format = '%Y')

corn_price_received_pivot = corn_price_received.pivot(index = 'year', columns = 'state', values = 'value')
corn_price_received_pivot.index = pd.to_datetime(corn_price_received_pivot.index, format = '%Y')

soy_price_received_pivot = soy_price_received.pivot(index = 'year', columns = 'state', values = 'value')
soy_price_received_pivot.index = pd.to_datetime(soy_price_received_pivot.index, format = '%Y')
```

```python
# Same pivoting for the economic data sets

# Reshaping dataframes to be able to pivot
employment_reshaped = pd.melt(employment, id_vars = 'state', var_name = 'year', value_name = 'value')
employment_reshaped['year'] = pd.to_datetime(employment_reshaped['year'], format = '%Y')
employment_pivot = employment_reshaped.pivot(index = 'year', columns = 'state', values = 'value')

# Same process for the other economic data sets
personal_income_reshaped = pd.melt(personal_income, id_vars = 'state', var_name = 'year', value_name = 'value')
personal_income_reshaped['year'] = pd.to_datetime(personal_income_reshaped['year'], format = '%Y')
personal_income_pivot = personal_income_reshaped.pivot(index = 'year', columns = 'state', values = 'value')
```

```python
# Instantiating figure
plt.figure(figsize = (10, 6))

# Plotting lines for each state
for state in employment_pivot.sum().nlargest(20).index:
    plt.plot(employment_pivot.index, employment_pivot[state] / 1_000, label = state)

plt.title('Rural Employment Trends by State From 2000-2023', size = 20)
plt.xlabel('Year', size = 20)
plt.ylabel('Employed Population (Thousands)', size = 15)
plt.legend(title = 'State', bbox_to_anchor = (1.2, 1), loc = 'upper right', fontsize = 'small')

plt.grid(True)
plt.xticks(rotation = 45)

plt.show();
```

```python
# Extracting the poverty percentages by state and year from poverty_levels
poverty_percent = poverty_levels.pivot(index = 'year', columns = 'state', values = 'percent_in_poverty')

# Instantiating figure
plt.figure(figsize = (10, 6))

# Plotting lines for each state
for state in poverty_percent.sum().nlargest(10).index:
    plt.plot(poverty_percent.index, poverty_percent[state], label = state)

plt.title('Top 10 Poverty Rates by State From 2000-2023', size = 20)
plt.xlabel('Year', size = 20)
plt.ylabel('Poverty Rate (%)', size = 15)
plt.legend(title = 'State', bbox_to_anchor = (1.22, 1), loc = 'upper right')

plt.grid(True)
plt.xticks(rotation = 45)

plt.show();
```

```python
# Taking aggregate economic data grouped by producing and non producing states
non_corn_producing_states = ['Alaska', 'Connecticut', 'Hawaii', 'Maine', 'Massachusetts', 'Nevada', 'New Hampshire', 'Rhode Island', 'Vermont']

# Mean Personal Income
income_non_corn_producing = round(personal_income_pivot[non_corn_producing_states].mean().mean())
income_corn_producing = round(personal_income_pivot.drop(columns = non_corn_producing_states).mean().mean())

# Poverty Level
poverty_non_corn_producing = round(poverty_percent[non_corn_producing_states].mean().mean(), 2)
poverty_corn_producing = round(poverty_percent.drop(columns = non_corn_producing_states).mean().mean(), 2)

# Employment
employment_non_corn_producing = round(employment_pivot[non_corn_producing_states].mean().mean(), 2)
employment_corn_producing = round(employment_pivot.drop(columns = non_corn_producing_states).mean().mean(), 2)
```

```python
# Comparing economic aggregates of corn producing and non corn producing states via bar graph

plt.figure(figsize = (10, 6))

categories = ['States That Produce Corn', 'States That Don\'t Produce Corn']

plt.bar(categories, [income_corn_producing / 1_000_000, income_non_corn_producing / 1_000_000],
       color = ['goldenrod', 'chocolate'],
       edgecolor = 'black')

plt.title('Average Total Income', fontsize = 25)
plt.ylabel('Total Income of State in USD (Millions)', fontsize = 15)
plt.xticks(fontsize = 15)

plt.show();
```

```python
# Comparing economic aggregates of corn producing and non corn producing states via bar graph

plt.figure(figsize = (10, 6))

categories = ['States That Produce Corn', 'States That Don\'t Produce Corn']

plt.bar(categories, [poverty_corn_producing, poverty_non_corn_producing],
       color = ['goldenrod', 'chocolate'],
       edgecolor = 'black')

plt.title('Average Poverty Rate', fontsize = 25)
plt.ylabel('Percentage of Population in Poverty', fontsize = 15)
plt.xticks(fontsize = 15)

plt.show();
```

```python
# Comparing economic aggregates of corn producing and non corn producing states via bar graph

plt.figure(figsize = (10, 6))

categories = ['States That Produce Corn', 'States That Don\'t Produce Corn']

plt.bar(categories, [employment_corn_producing, employment_non_corn_producing],
       color = ['goldenrod', 'chocolate'],
       edgecolor = 'black')

plt.title('Average Employed Population', fontsize = 25)
plt.ylabel('Population Employed', fontsize = 15)
plt.xticks(fontsize = 15)

plt.show();
```
